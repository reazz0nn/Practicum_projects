## 3. –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

–ö–æ–ª–ª–µ–≥–∏ —Ö–æ—Ç—è—Ç –∏–∑—É—á–∏—Ç—å –∏—Å—Ç–æ—Ä–∏—é –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –≤ –Ω–∞—á–∞–ª–µ XXI –≤–µ–∫–∞, –∏ –∏—Ö –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç –ø–µ—Ä–∏–æ–¥ —Å 2000 –ø–æ 2013 –≥–æ–¥ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. –û—Ç–±–µ—Ä–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —ç—Ç–æ–º—É –ø–æ–∫–∞–∑–∞—Ç–µ–ª—é. –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ –Ω–æ–≤—ã–π —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ, –Ω–∞–ø—Ä–∏–º–µ—Ä `df_actual`.

```python
#–û—Ç–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –∑–∞ –ø–µ—Ä–∏–æ–¥ —Å 2000 –ø–æ 2013 –≥–æ–¥
df_actual = df[(df['year_of_release']>=2000) & (df['year_of_release']<=2013)].copy() 
#–ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞–∑–º–µ—Ä –∏ –ø–µ—Ä–≤—ã–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫ –Ω–æ–≤–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
display(df_actual.shape)
display(df_actual.head())
```

## 4. –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
    
–ü—Ä–æ–≤–µ–¥–∏—Ç–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—é –¥–∞–Ω–Ω—ã—Ö:
- –†–∞–∑–¥–µ–ª–∏—Ç–µ –≤—Å–µ –∏–≥—Ä—ã –ø–æ –æ—Ü–µ–Ω–∫–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –≤—ã–¥–µ–ª–∏—Ç–µ —Ç–∞–∫–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: –≤—ã—Å–æ–∫–∞—è –æ—Ü–µ–Ω–∫–∞ (–æ—Ç 8 –¥–æ 10 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ), —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ (–æ—Ç 3 –¥–æ 8, –Ω–µ –≤–∫–ª—é—á–∞—è –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞) –∏ –Ω–∏–∑–∫–∞—è –æ—Ü–µ–Ω–∫–∞ (–æ—Ç 0 –¥–æ 3, –Ω–µ –≤–∫–ª—é—á–∞—è –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞).

```python
#–ó–∞–¥–∞–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤
bins = [-5,0,3,8,10]
#–ó–∞–¥–∞–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –æ—Ü–µ–Ω–∏–≤–∞–Ω–∏—è
labels = ['–æ—Ç—Å—É—Ç—Å–≤—É–µ—Ç –æ—Ü–µ–Ω–∫–∞','–Ω–∏–∑–∫–∞—è –æ—Ü–µ–Ω–∫–∞', '—Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞', '–≤—ã—Å–æ–∫–∞—è –æ—Ü–µ–Ω–∫–∞']
#–°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü 'user_rating'
df['user_rating'] = pd.cut(df['user_score'], bins=bins,labels=labels,right=False,include_lowest=True)
#–ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –æ—Ü–µ–Ω–∫–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –≤–µ–¥–µ–º –ø–æ–¥—Å—á–µ—Ç –∏–≥—Ä
group = df.groupby('user_rating',observed=False).agg({'name':'count', 'user_score':'mean'}).reset_index()
group.columns = ['user_rating','amount_of_games','avg_count']
display(group)
```

- –†–∞–∑–¥–µ–ª–∏—Ç–µ –≤—Å–µ –∏–≥—Ä—ã –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ –≤—ã–¥–µ–ª–∏—Ç–µ —Ç–∞–∫–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: –≤—ã—Å–æ–∫–∞—è –æ—Ü–µ–Ω–∫–∞ (–æ—Ç 80 –¥–æ 100 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ), —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ (–æ—Ç 30 –¥–æ 80, –Ω–µ –≤–∫–ª—é—á–∞—è –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞) –∏ –Ω–∏–∑–∫–∞—è –æ—Ü–µ–Ω–∫–∞ (–æ—Ç 0 –¥–æ 30, –Ω–µ –≤–∫–ª—é—á–∞—è –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞).

```python
#–ó–∞–¥–∞–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤
bins = [-5,0,30,80,100]
#–ó–∞–¥–∞–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –æ—Ü–µ–Ω–∏–≤–∞–Ω–∏—è
labels = ['–æ—Ç—Å—É—Ç—Å–≤—É–µ—Ç –æ—Ü–µ–Ω–∫–∞','–Ω–∏–∑–∫–∞—è –æ—Ü–µ–Ω–∫–∞', '—Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞', '–≤—ã—Å–æ–∫–∞—è –æ—Ü–µ–Ω–∫–∞']
#–°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü 'critic_rating'
df['critic_rating'] = pd.cut(df['critic_score'], bins=bins,labels=labels,right=False)
#–ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ –≤–µ–¥–µ–º –ø–æ–¥—Å—á–µ—Ç –∏–≥—Ä
group = df.groupby('critic_rating',observed=False).agg({'name':'count','critic_score':'mean'}).reset_index()
group.columns = ['critic_rating','amount_of_games','avg_count']
display(group)
```
